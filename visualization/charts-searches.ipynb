```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Pesquisa da CNT no mês de Março
https://especiais.gazetadopovo.com.br/eleicoes/2018/wp-content/uploads/2018/03/cntmda-201803.pdf

```python
plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Lula','Jair Bolsonaro','Ciro Gomes','Geraldo Alckmin','Alvaro Dias','Marina Silva','Michel Temer','Outras candidatos')
y_pos = np.arange(len(people))
performance = [18.6,12.3,1.7,1.4,1.2,1.2,0.4,3.1]

ax.barh(y_pos, performance, align='center',
        color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Intenções de voto em %')
ax.set_title('Intenção de voto para presidente 2018 no mês de Março')
plt.text(0.42, -0.05, "Fonte: CNT", fontsize=12, transform=plt.gcf().transFigure)

plt.show()
```
